## 1.4 Data cleaning

Our cleaning pipeline is simple and conservative:

1. **Forward fill** small gaps (weekends, holidays where some markets are open and others are not)
2. **Backward fill** any remaining NaN at the start of the series
3. **Drop** rows where more than 50% of assets have NaN (complete market closures)
4. **Outlier check**: flag any daily return exceeding +/- 50% (likely data error, not a real move)

```python
# ============================================================
# Step 1: Forward fill, then backward fill for leading NaNs
# This is standard practice for multi-asset datasets with
# different trading calendars (cf. Lecture 1 on data handling)
# ============================================================
rows_before = len(df)
nan_before = df.isnull().sum().sum()

df = df.ffill()   # forward fill: carry last known price
df = df.bfill()   # backward fill: handle start-of-series gaps

# Drop rows where too many assets are still missing
threshold = df.shape[1] * 0.5  # at least 50% of columns must be non-NaN
df = df.dropna(thresh=int(threshold))

# Drop any remaining NaN rows (edge cases)
df = df.dropna()

nan_after = df.isnull().sum().sum()
rows_after = len(df)

print(f'NaN values: {nan_before} -> {nan_after}')
print(f'Rows: {rows_before} -> {rows_after} (dropped {rows_before - rows_after})')
print(f'Final shape: {df.shape}')
```

```python
# ============================================================
# Step 2: Outlier detection on daily returns
# Flag any single-day return > 50% as suspicious
# (Oil did drop ~30% in a day during COVID, so we keep that)
# ============================================================
returns = df.pct_change().dropna()

extreme_threshold = 0.50  # 50%
outliers = (returns.abs() > extreme_threshold)
n_outliers = outliers.sum().sum()

if n_outliers > 0:
    print(f'Found {n_outliers} extreme returns (> +/-50%):')
    for col in returns.columns:
        extreme_dates = returns.index[outliers[col]]
        for d in extreme_dates:
            print(f'  {col} on {d.strftime("%Y-%m-%d")}: {returns.loc[d, col]:.2%}')
    print('\nThese are likely data errors. We will cap them at +/-50%.')
    # Cap extreme returns and reconstruct prices
    returns_capped = returns.clip(-extreme_threshold, extreme_threshold)
    df_clean = df.iloc[0:1].copy()  # keep first row
    for i in range(len(returns_capped)):
        new_row = df_clean.iloc[-1] * (1 + returns_capped.iloc[i])
        df_clean = pd.concat([df_clean, new_row.to_frame().T])
    df_clean.index = df.index
    df = df_clean
else:
    print('No extreme returns detected (all daily returns within +/-50%).')
    print('The COVID crash in March 2020 had ~30% drops, which we keep as legitimate.')
```

<div style="background-color: #ffebee; border-left: 4px solid #c62828; padding: 10px; margin: 10px 0;">
<b>Important Note</b><br>
We intentionally keep the March 2020 COVID crash and the Feb 2022 Russia-Ukraine spike in the data.
These are real market events, not data errors, and they are crucial for testing our strategy in
extreme conditions. The 50% threshold only catches genuine data glitches.
</div>

```python
# ============================================================
# Verify the date range and continuity
# ============================================================

# Check for large gaps in dates (more than 5 business days)
date_diffs = pd.Series(df.index).diff().dt.days
large_gaps = date_diffs[date_diffs > 5]

if len(large_gaps) > 0:
    print(f'Found {len(large_gaps)} gaps larger than 5 calendar days:')
    for idx in large_gaps.index:
        print(f'  {df.index[idx-1].strftime("%Y-%m-%d")} -> {df.index[idx].strftime("%Y-%m-%d")} ({int(date_diffs[idx])} days)')
else:
    print('No suspicious gaps in the time series. Data looks continuous.')

print(f'\nFinal dataset: {df.shape[0]} trading days, {df.shape[1]} assets')
print(f'From {df.index[0].strftime("%Y-%m-%d")} to {df.index[-1].strftime("%Y-%m-%d")}')
```

## 1.5 Descriptive statistics

```python
# ============================================================
# Full descriptive stats on price levels
# ============================================================
desc = df.describe().round(2)
desc
```

```python
# ============================================================
# Return statistics (annualized)
# These are the stylized facts we'll refer to later
# (cf. Lecture 1 on stylized facts of commodity prices)
# ============================================================
returns = df.pct_change().dropna()

return_stats = pd.DataFrame({
    'Ann. Return (%)': (returns.mean() * 252 * 100).round(2),
    'Ann. Vol (%)': (returns.std() * np.sqrt(252) * 100).round(2),
    'Sharpe (rf=0)': ((returns.mean() / returns.std()) * np.sqrt(252)).round(3),
    'Skewness': returns.skew().round(3),
    'Kurtosis': returns.kurtosis().round(3),
    'Min Daily (%)': (returns.min() * 100).round(2),
    'Max Daily (%)': (returns.max() * 100).round(2)
}).sort_values('Sharpe (rf=0)', ascending=False)

return_stats
```

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 1, Slides 10-18)</b><br>
The stylized facts of commodity returns include: negative skewness (crash risk), excess kurtosis
(fat tails, more extreme moves than a Gaussian distribution), and volatility clustering (large moves
tend to follow large moves). We check all three in the table above.
</div>

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
A few things stand out from the return statistics. First, all assets show excess kurtosis well above 0,
confirming that energy returns have much fatter tails than the normal distribution. This is
consistent with what we learned in Lecture 1 about commodity stylized facts. Second, most
assets have negative or near-zero skewness, reflecting the asymmetry we expect in equity-like
instruments (crashes are faster than rallies). The annualized volatility of Brent (<code>BZ=F</code>) is typically
higher than that of diversified ETFs like <code>XLE</code>, which makes sense since the ETF pools exposure
across many companies. The Sharpe ratios give us a first idea of which assets have the best
risk-adjusted performance over the full sample, but keep in mind this includes both in-sample
and out-of-sample periods mixed together. We will split them properly later.
</div>

```python
# ============================================================
# Define in-sample / out-of-sample split
# In-sample: 2019-01 to 2023-12 (5 years, for calibration)
# Out-of-sample: 2024-01 to 2026-03 (2+ years, for testing)
# ============================================================
SPLIT_DATE = '2024-01-01'

df_is = df[df.index < SPLIT_DATE]   # in-sample
df_oos = df[df.index >= SPLIT_DATE]  # out-of-sample

print(f'In-sample:     {df_is.index[0].strftime("%Y-%m-%d")} to {df_is.index[-1].strftime("%Y-%m-%d")} ({len(df_is)} days)')
print(f'Out-of-sample: {df_oos.index[0].strftime("%Y-%m-%d")} to {df_oos.index[-1].strftime("%Y-%m-%d")} ({len(df_oos)} days)')
```

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 2)</b><br>
Palazzi uses an in-sample / out-of-sample split to avoid look-ahead bias. We follow the same
principle: all parameter optimization (hedge ratio, lookback window, volatility threshold) is done
on the in-sample period only. The out-of-sample period is reserved for genuine backtesting.
</div>

```python
# ============================================================
# Quick sanity check: plot Brent price to see if data looks right
# We should see: 2020 COVID crash, 2022 Russia-Ukraine spike
# ============================================================
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(df['BZ=F'], color='#1565c0', linewidth=1.2, label='Brent Crude (BZ=F)')

# Mark key events
ax.axvline(pd.Timestamp('2020-03-09'), color='red', linestyle='--', alpha=0.7, label='COVID crash (Mar 2020)')
ax.axvline(pd.Timestamp('2022-02-24'), color='orange', linestyle='--', alpha=0.7, label='Russia-Ukraine (Feb 2022)')
ax.axvline(pd.Timestamp(SPLIT_DATE), color='green', linestyle='--', alpha=0.7, label=f'Train/Test split ({SPLIT_DATE})')

ax.set_title('Brent Crude Oil Futures: Sanity Check', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD/barrel)')
ax.legend(loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">

The Brent price series looks correct. We can clearly see the COVID crash in March 2020 where

Russia's invasion of Ukraine in February 2022, and the gradual normalization since then. The
in-sample period (left of the green line) captures all of these extreme regimes, which is good
for calibrating a strategy that needs to handle both calm and turbulent markets.
