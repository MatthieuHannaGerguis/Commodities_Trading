## 1.4 Data cleaning

Our cleaning pipeline:

1. **Forward fill** to handle small gaps (holiday mismatches between exchanges)
2. **Backward fill** for any NaN at the very start of a series
3. **Outlier check**: flag daily returns exceeding +/-50% (data errors, not real moves)
4. **Date continuity check**: detect any suspicious gaps in the time series

```python
# Step 1: Fill missing values (cf. Lecture 1 on data handling)
nan_before = df.isnull().sum().sum()
df = df.ffill()
df = df.bfill()
df = df.dropna()
nan_after = df.isnull().sum().sum()
print(f'NaN values: {nan_before} -> {nan_after}')
print(f'Dataset shape after cleaning: {df.shape}')
```

```python
# Step 2: Outlier detection (flag daily returns > +/-50%)
returns_raw = df.pct_change().dropna()
extreme_threshold = 0.50
outliers = (returns_raw.abs() > extreme_threshold)
n_outliers = outliers.sum().sum()

if n_outliers > 0:
    print(f'WARNING: {n_outliers} extreme daily returns detected:')
    for col in returns_raw.columns:
        for d in returns_raw.index[outliers[col]]:
            print(f'  {col} on {d.strftime("%Y-%m-%d")}: {returns_raw.loc[d, col]:+.2%}')
    print('Capping at +/-50%.')
    returns_capped = returns_raw.clip(-extreme_threshold, extreme_threshold)
    df_reconstructed = (1 + returns_capped).cumprod() * df.iloc[0]
    df_reconstructed.index = returns_capped.index
    df = pd.concat([df.iloc[:1], df_reconstructed])
else:
    print('No extreme returns detected (all daily moves within +/-50%).')
    print('COVID crash (Mar 2020) had ~30% drops, which we keep as real events.')
```

<div style="background-color: #ffebee; border-left: 4px solid #c62828; padding: 10px; margin: 10px 0;">
<b>Important Note</b><br>
We intentionally keep the March 2020 COVID crash and the February 2022 Russia-Ukraine
spike. These are real market events, not data errors. Our strategy needs to handle extreme
conditions, so removing them would defeat the purpose of a realistic backtest.
</div>

```python
# Step 3: Date continuity check
date_diffs = pd.Series(df.index).diff().dt.days
large_gaps = date_diffs[date_diffs > 5]
if len(large_gaps) > 0:
    print(f'Found {len(large_gaps)} gap(s) larger than 5 calendar days:')
    for idx in large_gaps.index:
        print(f'  {df.index[idx-1].strftime("%Y-%m-%d")} -> {df.index[idx].strftime("%Y-%m-%d")} ({int(date_diffs[idx])} days)')
    print('Likely long holiday weekends. No action needed.')
else:
    print('No suspicious gaps. Time series is continuous.')

print(f'\n=== CLEANING COMPLETE ===')
print(f'Final dataset: {df.shape[0]} trading days, {df.shape[1]} assets')
print(f'Period: {df.index[0].strftime("%Y-%m-%d")} to {df.index[-1].strftime("%Y-%m-%d")}')
```

## 1.5 In-sample / Out-of-sample split

```python
# cf. Palazzi (2025) Section 2 on avoiding look-ahead bias
df_is  = df.loc[df.index < SPLIT_DATE].copy()
df_oos = df.loc[df.index >= SPLIT_DATE].copy()

print(f'In-sample:     {df_is.index[0].strftime("%Y-%m-%d")} to {df_is.index[-1].strftime("%Y-%m-%d")}  ({len(df_is)} days)')
print(f'Out-of-sample: {df_oos.index[0].strftime("%Y-%m-%d")} to {df_oos.index[-1].strftime("%Y-%m-%d")}  ({len(df_oos)} days)')
print(f'\nRatio: {len(df_is)/len(df):.1%} in-sample, {len(df_oos)/len(df):.1%} out-of-sample')
```

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 2)</b><br>
Palazzi uses a strict in-sample / out-of-sample protocol to avoid look-ahead bias. All
parameter optimization (hedge ratio, lookback window, z-score thresholds, volatility filter)
is calibrated on in-sample data only. The out-of-sample period is a true blind test.
</div>

## 2.2 Stylized facts: return distributions, autocorrelation, volatility clustering

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 1, Slides 10-18 & Geman 2005, Table 3.1)</b><br>
Commodity returns exhibit well-known stylized facts: mean-reversion in prices (not in returns),
fat tails (excess kurtosis), negative skewness, volatility clustering, and near-zero autocorrelation
in returns but significant autocorrelation in squared returns. Geman (2005) reports Brent kurtosis
of about 6.08 and skewness of 0.16 for the 1993-2000 period. We check whether our 2019-2026
sample shows similar patterns.
</div>

```python
# Return distribution for Brent: histogram + QQ plot
brent_ret = returns['BZ=F']

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: histogram with normal overlay
axes[0].hist(brent_ret, bins=80, density=True, alpha=0.6, color='#1565c0',
             edgecolor='white', label='Brent daily returns')
x_range = np.linspace(brent_ret.min(), brent_ret.max(), 200)
normal_pdf = stats.norm.pdf(x_range, brent_ret.mean(), brent_ret.std())
axes[0].plot(x_range, normal_pdf, 'r-', linewidth=2, label='Normal fit')
axes[0].set_title('Brent Returns: Histogram vs. Normal Distribution', fontsize=13)
axes[0].set_xlabel('Daily Return')
axes[0].set_ylabel('Density')
axes[0].legend()

# Right: QQ-plot
stats.probplot(brent_ret, dist='norm', plot=axes[1])
axes[1].set_title('Brent Returns: Q-Q Plot vs. Normal', fontsize=13)
axes[1].get_lines()[0].set_markerfacecolor('#1565c0')
axes[1].get_lines()[0].set_markersize(3)

plt.tight_layout()
plt.show()

jb_stat, jb_pval = stats.jarque_bera(brent_ret)
print(f'Jarque-Bera test: statistic = {jb_stat:.2f}, p-value = {jb_pval:.2e}')
print(f'Skewness = {brent_ret.skew():.3f}, Kurtosis = {brent_ret.kurtosis():.3f}')
if jb_pval < 0.01:
    print('=> Normality rejected at 1% level. Returns have fat tails.')
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The histogram and QQ-plot confirm what we expected: Brent returns are not normally distributed.
The tails are much heavier than a Gaussian (visible on the QQ-plot where extreme observations
deviate sharply from the diagonal). The Jarque-Bera test rejects normality with a tiny p-value.
This is consistent with Geman (2005, Table 3.1) who reports similar patterns for Brent over
1993-2000. The practical implication is that any strategy assuming normal returns will
underestimate tail risk. This motivates our GARCH volatility filter later in Section 4.
</div>

```python
# Return distributions for all assets: box plot
fig, ax = plt.subplots(figsize=(16, 6))
returns_plot = returns[col_order]
bp = ax.boxplot([returns_plot[col].dropna() for col in returns_plot.columns],
                labels=returns_plot.columns, patch_artist=True, showfliers=True,
                flierprops={'markersize': 2, 'alpha': 0.4})
colors_map = {'BZ=F': '#333333'}
for t in etfs: colors_map[t] = '#1565c0'
for t in equities: colors_map[t] = '#ff8f00'
for patch, col in zip(bp['boxes'], returns_plot.columns):
    patch.set_facecolor(colors_map.get(col, 'gray'))
    patch.set_alpha(0.6)
ax.set_title('Daily Return Distributions by Asset', fontsize=14)
ax.set_ylabel('Daily Return')
ax.axhline(0, color='black', linewidth=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The box plot confirms that all assets exhibit fat tails (many outliers beyond the whiskers).
PBR (Petrobras) and WDS (Woodside) have the widest interquartile ranges, reflecting additional
country-specific risk from Brazil and Australia. The ETFs (blue boxes) have tighter distributions,
as expected from diversified instruments. This variation in volatility levels across assets needs
to be accounted for when constructing pairs: we want assets with similar volatility profiles,
or we need to adjust position sizes accordingly.
</div>

# 3. Stationarity and Cointegration Tests

The logic here follows the three-step methodology from Lecture 3 (Slides 13-17) and
Palazzi (2025, Section 3): first confirm that each price series is non-stationary (I(1)), then
test for cointegration between Brent and each equity, and finally select the best pair for
our trading strategy. Cointegration is the statistical backbone of pairs trading: it tells us
that two series share a long-run equilibrium, and any deviation from that equilibrium is
temporary and should revert. Without cointegration, pairs trading is just gambling on
correlations that might break down at any time.

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 3, Slides 13-17 & Lecture 1, Slides 58-59)</b><br>
The procedure from the course is: (1) check if each series is I(1) using ADF and KPSS tests,
(2) test for cointegration using Engle-Granger (two-step) or Johansen (trace + max eigenvalue),
(3) if cointegrated, estimate the VECM for the long-run equilibrium. For the crack spread case
study (Lecture 3, Slide 36), the professor shows exactly this approach: individual series are I(1),
but the spread is I(0), confirming cointegration.
</div>

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 3.1) & GitHub</b><br>
Palazzi uses Engle-Granger cointegration to select the best pair from a universe of 10
cryptocurrencies. The pair with the lowest cointegration p-value is selected for trading. All
tests are performed on the in-sample period only, to avoid look-ahead bias. We follow the exact
same protocol, adapted to our 14 energy assets.
</div>

## 3.1 ADF and KPSS tests on price levels

We run two complementary tests on each price series (in log-levels, in-sample only):

**ADF test** (Augmented Dickey-Fuller): H0 = unit root (non-stationary). If p-value > 0.05, we cannot reject H0, meaning the series is likely non-stationary. This is what we expect for price levels.

**KPSS test**: H0 = stationarity. If p-value < 0.05, we reject stationarity. KPSS and ADF are complementary because they have opposite null hypotheses. If ADF says "cannot reject unit root" AND KPSS says "reject stationarity", we have strong evidence that the series is I(1).

```python
# ============================================================
# ADF and KPSS tests on LOG PRICE LEVELS (in-sample only)
# cf. Lecture 1, Slide 58 (ADF) and Lecture 3, Slide 13
# We expect: ADF fails to reject H0 (non-stationary)
#            KPSS rejects H0 (not stationary)
# => Series is I(1), prerequisite for cointegration
# ============================================================
log_prices_is = np.log(df_is)  # log-prices for stationarity tests

stationarity_results = []

for col in df_is.columns:
    series = log_prices_is[col].dropna()

    # ADF test (H0: unit root / non-stationary)
    adf_stat, adf_pval, adf_lags, _, _, _ = adfuller(series, autolag='AIC')

    # KPSS test (H0: stationary)
    kpss_stat, kpss_pval, kpss_lags, _ = kpss(series, regression='c', nlags='auto')

    # Determine integration order
    if adf_pval > 0.05 and kpss_pval < 0.05:
        order = 'I(1)'  # non-stationary, good for cointegration
    elif adf_pval <= 0.05 and kpss_pval >= 0.05:
        order = 'I(0)'  # already stationary
    else:
        order = '???'    # ambiguous

    cat = categories.get(col, '?')
    stationarity_results.append({
        'Ticker': col,
        'Category': cat,
        'ADF Stat': round(adf_stat, 3),
        'ADF p-val': round(adf_pval, 4),
        'KPSS Stat': round(kpss_stat, 3),
        'KPSS p-val': round(kpss_pval, 4),
        'Order': order
    })

stationarity_df = pd.DataFrame(stationarity_results).set_index('Ticker')
print('Stationarity tests on LOG PRICE LEVELS (in-sample, 2019-2023):')
print('ADF H0: unit root (non-stationary) | KPSS H0: stationary')
print()
stationarity_df
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
As expected, all price series in log-levels are I(1): the ADF test fails to reject the unit root
hypothesis (high p-values), and KPSS rejects stationarity (low p-values). This is the standard
result for financial price series and the prerequisite for cointegration analysis. If any series
were already stationary (I(0)), it would not be suitable for pairs trading via cointegration
because there would be no "common stochastic trend" to exploit.
</div>

```python
# ============================================================
# Quick sanity check: ADF on RETURNS (should be stationary)
# If prices are I(1), returns should be I(0)
# ============================================================
returns_is = log_prices_is.diff().dropna()

print('ADF tests on LOG-RETURNS (should be stationary, i.e. p-val < 0.05):')
for col in df_is.columns:
    adf_stat, adf_pval, _, _, _, _ = adfuller(returns_is[col].dropna(), autolag='AIC')
    status = 'I(0) OK' if adf_pval < 0.05 else 'NOT stationary (!)'
    print(f'  {col:8s}: ADF stat = {adf_stat:.3f}, p-val = {adf_pval:.4f}  => {status}')
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
All log-return series are stationary (ADF p-values essentially 0), confirming that our price
series are integrated of order 1. This double-check (levels are I(1), returns are I(0)) is good
practice and follows the methodology outlined in Lecture 3 (Slide 13).
</div>

## 3.2 Engle-Granger cointegration (all pairs with Brent)

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 3, Slides 15-17)</b><br>
The Engle-Granger two-step procedure: (1) regress Y on X via OLS to get the residuals,
(2) test the residuals for stationarity. If the residuals are stationary (I(0)), then Y and X are
cointegrated. The <code>statsmodels.tsa.stattools.coint</code> function does both steps automatically.
We apply this to every pair (Brent vs. each of the other 13 assets) using in-sample data only.
</div>

```python
# ============================================================
# Engle-Granger cointegration test: Brent vs. every other asset
# Using LOG PRICES on the IN-SAMPLE period only
# cf. Lecture 3, Slides 15-17 and Palazzi (2025) Section 3.1
# ============================================================
brent_log = log_prices_is['BZ=F']

coint_results = []

for col in df_is.columns:
    if col == 'BZ=F':
        continue

    other_log = log_prices_is[col].dropna()

    # Engle-Granger test (statsmodels implementation)
    # Returns: test statistic, p-value, critical values
    coint_stat, coint_pval, coint_crit = coint(brent_log, other_log)

    cat = categories.get(col, '?')
    coint_results.append({
        'Pair': f'Brent vs {col}',
        'Category': cat,
        'EG Stat': round(coint_stat, 3),
        'p-value': round(coint_pval, 4),
        '1% CV': round(coint_crit[0], 3),
        '5% CV': round(coint_crit[1], 3),
        '10% CV': round(coint_crit[2], 3),
        'Cointegrated (5%)': 'Yes' if coint_pval < 0.05 else 'No'
    })

coint_df = pd.DataFrame(coint_results).sort_values('p-value')
print('Engle-Granger Cointegration Tests: Brent vs. All Assets (In-Sample)')
print('H0: No cointegration | Reject if p-value < 0.05')
print()
coint_df
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The cointegration results show which assets share a genuine long-run equilibrium with Brent.
Not all correlated assets are cointegrated. Correlation measures short-term co-movement, while
cointegration captures a long-run relationship that prevents the two series from drifting apart
indefinitely. The pairs with the lowest p-values are our best candidates for the trading strategy.
We need the selected pair to make economic sense too: a statistical relationship without
an economic rationale is fragile and likely to break down out-of-sample.
</div>

```python
# ============================================================
# Visualization: cointegration p-values ranked
# ============================================================
fig, ax = plt.subplots(figsize=(10, 6))

pairs = coint_df['Pair'].str.replace('Brent vs ', '')
pvals = coint_df['p-value'].values
colors = ['#2e7d32' if p < 0.05 else '#c62828' if p > 0.10 else '#f57f17' for p in pvals]

bars = ax.barh(range(len(pairs)), pvals, color=colors, edgecolor='white')
ax.set_yticks(range(len(pairs)))
ax.set_yticklabels(pairs)
ax.set_xlabel('Cointegration p-value (Engle-Granger)')
ax.set_title('Cointegration with Brent: All Pairs (In-Sample)', fontsize=14)
ax.invert_yaxis()

# Significance lines
ax.axvline(0.05, color='green', linestyle='--', alpha=0.7, label='5% significance')
ax.axvline(0.10, color='orange', linestyle='--', alpha=0.5, label='10% significance')

for i, v in enumerate(pvals):
    ax.text(v + 0.005, i, f'{v:.4f}', va='center', fontsize=9)

ax.legend(fontsize=9)
plt.tight_layout()
plt.show()

# Count significant pairs
n_sig_5 = (coint_df['p-value'] < 0.05).sum()
n_sig_10 = (coint_df['p-value'] < 0.10).sum()
print(f'Pairs cointegrated at 5%: {n_sig_5}')
print(f'Pairs cointegrated at 10%: {n_sig_10}')
```

## 3.3 Best pair selection

```python
# ============================================================
# Select the best pair for trading
# Criteria (following Palazzi 2025):
# 1. Lowest cointegration p-value
# 2. Economic relationship makes sense
# 3. Good data quality (no excessive gaps)
# ============================================================
best_pair = coint_df.iloc[0]  # lowest p-value

best_ticker = best_pair['Pair'].replace('Brent vs ', '')
best_pval = best_pair['p-value']
best_category = best_pair['Category']

print(f'=== BEST PAIR SELECTED ===')
print(f'Pair: Brent (BZ=F) vs {best_ticker}')
print(f'Category: {best_category}')
print(f'Cointegration p-value: {best_pval:.6f}')
print(f'')

# Also show top 3 for discussion
print('Top 3 candidates:')
for i, row in coint_df.head(3).iterrows():
    ticker = row['Pair'].replace('Brent vs ', '')
    print(f'  {row["Pair"]}: p-value = {row["p-value"]:.4f} ({row["Category"]})')
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
We select the pair with the lowest Engle-Granger p-value. This is the same selection criterion
used by Palazzi (2025): purely statistical, based on the strength of the cointegration relationship.
The economic rationale is also important: the selected equity should be an oil company whose
stock price is fundamentally driven by crude oil prices, creating a natural long-run equilibrium.
European oil majors (SHEL, TTE, BP, ENI) are particularly good candidates because they have

```python
# ============================================================
# Visualize the selected pair: price levels over time
# ============================================================
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

# Panel 1: both prices (different scales, dual y-axis)
ax1 = axes[0]
ax2 = ax1.twinx()
l1 = ax1.plot(df['BZ=F'], color='#1565c0', linewidth=1.3, label='Brent (BZ=F)')
l2 = ax2.plot(df[best_ticker], color='#c62828', linewidth=1.3, label=best_ticker)
add_event_lines(ax1)
ax1.set_ylabel('Brent Price (USD)', color='#1565c0')
ax2.set_ylabel(f'{best_ticker} Price (USD)', color='#c62828')
lines = l1 + l2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, fontsize=9, loc='upper left')
ax1.set_title(f'Selected Pair: Brent vs {best_ticker} (Price Levels)', fontsize=13)

# Panel 2: log-price spread (visual check for mean-reversion)
log_brent = np.log(df['BZ=F'])
log_other = np.log(df[best_ticker])

# Quick OLS hedge ratio on in-sample for spread construction
from numpy.polynomial.polynomial import polyfit
b, a = np.polyfit(log_prices_is[best_ticker], log_prices_is['BZ=F'], 1)
spread = log_brent - b * log_other - a

axes[1].plot(spread, color='purple', linewidth=1)
axes[1].axhline(spread.mean(), color='black', linestyle='-', linewidth=0.8, label='Mean')
axes[1].axhline(0, color='gray', linestyle=':', alpha=0.5)
add_event_lines(axes[1])
axes[1].set_title(f'Log-Price Spread: Brent - {b:.2f} * {best_ticker} (visual check)', fontsize=13)
axes[1].set_ylabel('Spread')
axes[1].set_xlabel('Date')
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.show()
```

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 3, Slide 17: Gas-Oil Cointegration, Ohana 2006)</b><br>
The spread visualization follows the same logic as the Gas-Oil cointegration example from
Lecture 3. Ohana (2006) regresses log gas prices on log crude oil prices and finds
R-squared = 75.23% with stationary residuals (ADF p-value = 6%). Our approach is identical:
we regress log Brent on log equity price, and the spread should look mean-reverting.
If the spread wanders off and never comes back, the cointegration relationship has broken down.
</div>

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The spread plot should show a pattern that oscillates around its mean, which is the visual
signature of mean-reversion. Some deviations during COVID and Ukraine are expected (these
are regime breaks), but the spread should eventually return toward equilibrium. If it does,
this confirms that cointegration is working and our pairs trading strategy has a statistical
foundation to exploit.
</div>

```python
# ============================================================
# Confirm: ADF test on the spread (should be stationary)
# This is the residual-based test from Engle-Granger step 2
# ============================================================
spread_is = spread.loc[spread.index < SPLIT_DATE]
adf_spread, adf_spread_pval, _, _, _, _ = adfuller(spread_is.dropna(), autolag='AIC')

print(f'ADF test on the in-sample spread (Brent vs {best_ticker}):')
print(f'  ADF statistic: {adf_spread:.4f}')
print(f'  p-value: {adf_spread_pval:.6f}')
if adf_spread_pval < 0.05:
    print(f'  => Spread is STATIONARY (I(0)). Cointegration confirmed.')
    print(f'     Mean-reversion is supported. Pairs trading has a statistical basis.')
else:
    print(f'  => Spread is NOT stationary at 5%. Cointegration is weak.')
    print(f'     Pairs trading on this pair is riskier.')
```

<div style="background-color: #e8f5e9; border-left: 4px solid #2e7d32; padding: 10px; margin: 10px 0;">
<b>TD Reference (Tutorial #3, Kalman Filter & Tutorial #4, Schwartz-Smith)</b><br>
In TD3 and TD4 we worked with state-space models and the concept of mean-reversion speed
(half-life). The spread we just constructed is essentially a stationary process with some
mean-reversion speed kappa. We could estimate this kappa using the Schwartz-Smith framework
from TD4 (the short-term factor chi follows an OU process with mean-reversion). The half-life
of mean-reversion (ln(2)/kappa) gives us an idea of how long we should expect to wait for
the spread to revert, which informs our lookback period optimization in Section 4.
</div>

```python
# ============================================================
# Estimate half-life of mean-reversion for the spread
# cf. TD4 (Schwartz-Smith): chi_t follows OU process with speed kappa
# Half-life = ln(2) / kappa
# We estimate kappa from an AR(1) on the spread
# ============================================================
spread_is_clean = spread_is.dropna()
spread_lag = spread_is_clean.shift(1).dropna()
spread_diff = spread_is_clean.diff().dropna()

# Align
common_idx = spread_lag.index.intersection(spread_diff.index)
y = spread_diff.loc[common_idx].values
x = spread_lag.loc[common_idx].values

# OLS: delta_spread = theta * spread_lag + error
# theta should be negative (mean-reverting)
theta = np.polyfit(x, y, 1)[0]
half_life = -np.log(2) / theta if theta < 0 else np.nan

print(f'Mean-reversion estimation for the spread:')
print(f'  AR(1) coefficient (theta): {theta:.4f}')
if theta < 0:
    print(f'  Half-life of mean-reversion: {half_life:.1f} trading days (~{half_life/21:.1f} months)')
    print(f'  This gives us a starting point for lookback period optimization in Section 4.')
else:
    print(f'  WARNING: theta is positive, suggesting no mean-reversion.')
    print(f'  The pair might not be suitable for pairs trading.')
```

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 3.2)</b><br>
Palazzi uses a grid search over lookback windows of different lengths to optimize the z-score
computation. The half-life of mean-reversion estimated here gives us economic intuition for
which lookback periods to try: a lookback much shorter than the half-life will be too noisy,
while one much longer will miss regime changes. We use this as a guide (not a hard constraint)
for the grid search in Section 4.3.
</div>

```python
print('=== SECTION 3 COMPLETE ===')
print(f'Selected pair: Brent (BZ=F) vs {best_ticker}')
print(f'Cointegration p-value: {best_pval:.6f}')
print(f'Spread is stationary: {"Yes" if adf_spread_pval < 0.05 else "No (weak)"}')
if theta < 0:
    print(f'Half-life of mean-reversion: {half_life:.1f} days')
print(f'\nReady for Section 4: Optimized Pairs Trading Strategy.')
```

# 4. Optimized Pairs Trading Strategy (Palazzi methodology)

This is the core section of the notebook. We implement the full optimized pairs trading
strategy from Palazzi (2025), adapted from crypto to energy commodities. The key components,
as specified in the project guidelines (Tutorial 1, Slide 8) and the Palazzi GitHub, are:

1. **Hedge ratio estimation** via OLS on in-sample log-prices
2. **Spread construction** and rolling z-score computation
3. **Lookback period optimization** via grid search (in-sample Sharpe maximization)
4. **Trading signals** based on z-score thresholds
5. **Volatility filter** to inhibit entry during high-volatility regimes
6. **Adaptive trailing stop-loss** to cut losses dynamically
7. **Minimum holding period** of 5 days before any exit is allowed

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025) & GitHub Parameters</b><br>
We follow the exact function signature from the Palazzi GitHub (<code>analyze_pairs_trading</code>):<br>
<code>split_ratio = 0.75</code> (we use ~68/32, similar order of magnitude)<br>
<code>threshold_value = 0.7</code> (z-score entry threshold, optimized via grid search)<br>
<code>transaction_cost = 0.002</code> (we use 0.001 = 10bps, as required by Tutorial 1)<br>
<code>trailing_stop_factor = 0.025</code> (2.5% trailing stop)<br>
<code>min_holding_period = 5</code> (minimum 5 trading days before exit)<br>
<code>vol_lookback = 30</code> (30-day rolling window for volatility computation)<br>
<code>vol_threshold = 1.5</code> (inhibit entry when vol > 1.5x the rolling average vol)
</div>

<div style="background-color: #ffebee; border-left: 4px solid #c62828; padding: 10px; margin: 10px 0;">
<b>Important: Transaction Costs</b><br>
As required by the project guidelines (Tutorial 1, Slide 7), we apply <b>10 basis points</b>
per trade (entry or exit). Palazzi uses 20bps (0.002) for crypto markets, but our professor
specified 10bps for equities/commodities, so we follow the course requirement.
</div>

## 4.1 Hedge ratio estimation (in-sample)

```python
# ============================================================
# OLS hedge ratio on in-sample log-prices
# We regress log(Brent) on log(Equity) to find the hedge ratio
# cf. Lecture 3, Slide 17 (Gas-Oil cointegration, Ohana 2006)
# ============================================================
from sklearn.linear_model import LinearRegression

log_brent_is = np.log(df_is['BZ=F']).values.reshape(-1, 1)
log_equity_is = np.log(df_is[best_ticker]).values.reshape(-1, 1)

# OLS regression: log(Brent) = alpha + beta * log(Equity)
reg = LinearRegression().fit(log_equity_is, log_brent_is)
hedge_ratio = reg.coef_[0][0]
intercept = reg.intercept_[0]
r_squared = reg.score(log_equity_is, log_brent_is)

print(f'=== HEDGE RATIO ESTIMATION (In-Sample) ===')
print(f'Pair: Brent vs {best_ticker}')
print(f'Hedge ratio (beta): {hedge_ratio:.4f}')
print(f'Intercept (alpha): {intercept:.4f}')
print(f'R-squared: {r_squared:.4f} ({r_squared*100:.1f}%)')
print(f'')
print(f'Interpretation: a 1% move in {best_ticker} corresponds to a ~{hedge_ratio:.2f}%')
print(f'move in Brent, on average over the in-sample period.')
```

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 3, Slide 17)</b><br>
This OLS regression follows exactly the Ohana (2006) approach presented in Lecture 3: regress
the log spot price of one commodity on another, then test the residuals for stationarity. The
beta coefficient is our hedge ratio, and the residuals form the spread we trade on.
</div>

## 4.2 Spread construction

```python
# ============================================================
# Construct the spread on the FULL dataset
# spread_t = log(Brent_t) - beta * log(Equity_t) - alpha
# The hedge ratio and intercept are estimated on in-sample ONLY
# ============================================================
log_brent_full = np.log(df['BZ=F'])
log_equity_full = np.log(df[best_ticker])

spread_full = log_brent_full - hedge_ratio * log_equity_full - intercept
spread_full.name = 'Spread'

print(f'Spread constructed: log(BZ=F) - {hedge_ratio:.4f} * log({best_ticker}) - {intercept:.4f}')
print(f'Spread stats (full sample):')
print(f'  Mean: {spread_full.mean():.6f}')
print(f'  Std:  {spread_full.std():.6f}')
print(f'  Min:  {spread_full.min():.6f}')
print(f'  Max:  {spread_full.max():.6f}')
```

## 4.3 Lookback period optimization

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 3.2)</b><br>
The lookback period determines the window over which we compute the rolling mean and std
of the spread, and therefore the z-score. Palazzi optimizes this via grid search on the
in-sample period, maximizing the Sharpe ratio. We search over windows of 30, 60, 90, 120,
180, and 252 days (where 252 trading days ~ 1 year). The half-life estimated in Section 3
gives us economic intuition for what range to expect.
</div>

```python
# ============================================================
# Grid search over lookback windows (IN-SAMPLE ONLY)
# For each lookback, we compute z-scores, generate signals,
# and measure the in-sample Sharpe ratio of the simple
# pairs strategy (no vol filter, no trailing stop yet).
# Adapted from Palazzi (2025) GitHub: analyze_pairs_trading()
# ============================================================

lookback_grid = [30, 60, 90, 120, 180, 252]
threshold_grid = [0.5, 0.7, 1.0, 1.5, 2.0]  # z-score entry thresholds

spread_is = spread_full.loc[spread_full.index < SPLIT_DATE].copy()
returns_brent_is = returns.loc[returns.index < SPLIT_DATE, 'BZ=F']

grid_results = []

for lookback in lookback_grid:
    for threshold in threshold_grid:
        # Rolling z-score
        roll_mean = spread_is.rolling(lookback).mean()
        roll_std = spread_is.rolling(lookback).std()
        zscore = (spread_is - roll_mean) / roll_std
        zscore = zscore.dropna()

        # Simple trading signals (no vol filter, no stop-loss)
        position = pd.Series(0.0, index=zscore.index)
        for i in range(1, len(zscore)):
            prev_pos = position.iloc[i-1]
            z = zscore.iloc[i]

            if prev_pos == 0:
                if z < -threshold:
                    position.iloc[i] = 1.0   # long spread (buy Brent, sell equity)
                elif z > threshold:
                    position.iloc[i] = -1.0  # short spread
                else:
                    position.iloc[i] = 0.0
            elif prev_pos == 1.0:
                if z >= 0:
                    position.iloc[i] = 0.0   # exit at mean reversion
                else:
                    position.iloc[i] = 1.0   # hold
            elif prev_pos == -1.0:
                if z <= 0:
                    position.iloc[i] = 0.0
                else:
                    position.iloc[i] = -1.0

        # Compute returns of the strategy (spread returns)
        spread_ret = spread_is.diff()
        strat_ret = (position.shift(1) * spread_ret).dropna()

        # Apply transaction costs on position changes
        trades = position.diff().abs()
        strat_ret = strat_ret - trades.shift(1).reindex(strat_ret.index).fillna(0) * TC

        # Sharpe ratio (annualized)
        if strat_ret.std() > 0:
            sharpe = (strat_ret.mean() / strat_ret.std()) * np.sqrt(252)
        else:
            sharpe = 0.0

        n_trades = (trades > 0).sum()

        grid_results.append({
            'Lookback': lookback,
            'Threshold': threshold,
            'Sharpe': round(sharpe, 3),
            'Ann. Return (%)': round(strat_ret.mean() * 252 * 100, 2),
            'Ann. Vol (%)': round(strat_ret.std() * np.sqrt(252) * 100, 2),
            'Trades': int(n_trades)
        })

grid_df = pd.DataFrame(grid_results)
print('Grid Search Results (In-Sample, Simple Pairs Strategy):')
print(grid_df.sort_values('Sharpe', ascending=False).head(15).to_string(index=False))
```

```python
# ============================================================
# Select optimal lookback and threshold
# ============================================================
best_row = grid_df.loc[grid_df['Sharpe'].idxmax()]
BEST_LOOKBACK = int(best_row['Lookback'])
BEST_THRESHOLD = float(best_row['Threshold'])

print(f'\n=== OPTIMAL PARAMETERS (In-Sample) ===')
print(f'Best lookback: {BEST_LOOKBACK} days')
print(f'Best threshold: {BEST_THRESHOLD}')
print(f'In-sample Sharpe: {best_row["Sharpe"]:.3f}')
print(f'In-sample Ann. Return: {best_row["Ann. Return (%)"]:.2f}%')
print(f'Number of trades: {int(best_row["Trades"])}')
```

```python
# ============================================================
# Heatmap of grid search results
# ============================================================
pivot = grid_df.pivot(index='Lookback', columns='Threshold', values='Sharpe')

fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='RdYlGn', center=0,
            linewidths=0.5, ax=ax)
ax.set_title('Grid Search: In-Sample Sharpe Ratio by Lookback x Threshold', fontsize=13)
ax.set_xlabel('Z-Score Entry Threshold')
ax.set_ylabel('Lookback Period (days)')
plt.tight_layout()
plt.show()
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The heatmap shows how the Sharpe ratio varies with lookback and threshold. There is usually
a "sweet spot" where the lookback is long enough to capture the mean-reversion signal but
short enough to adapt to regime changes. Very short lookbacks (30 days) tend to be noisy and
generate too many trades (high turnover costs), while very long ones (252 days) may miss
structural breaks. The threshold also matters: too low (0.5) generates many false signals,
while too high (2.0) misses opportunities. The optimal combination is what we use going forward.
</div>

## 4.4 Trading signals (z-score)

```python
# ============================================================
# Compute z-score with optimal parameters on FULL dataset
# ============================================================
roll_mean = spread_full.rolling(BEST_LOOKBACK).mean()
roll_std = spread_full.rolling(BEST_LOOKBACK).std()
zscore_full = ((spread_full - roll_mean) / roll_std).dropna()

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

# Top: spread
axes[0].plot(spread_full, color='purple', linewidth=1)
axes[0].plot(roll_mean, color='black', linewidth=0.8, linestyle='--', label='Rolling mean')
axes[0].fill_between(roll_mean.index,
                      roll_mean - BEST_THRESHOLD * roll_std,
                      roll_mean + BEST_THRESHOLD * roll_std,
                      alpha=0.15, color='gray', label=f'+/- {BEST_THRESHOLD} std')
add_event_lines(axes[0])
axes[0].set_title(f'Spread: Brent vs {best_ticker} (Lookback = {BEST_LOOKBACK} days)', fontsize=13)
axes[0].set_ylabel('Spread')
axes[0].legend(fontsize=8)

# Bottom: z-score with entry/exit zones
axes[1].plot(zscore_full, color='navy', linewidth=0.8)
axes[1].axhline(BEST_THRESHOLD, color='red', linestyle='--', alpha=0.7, label=f'Short entry (+{BEST_THRESHOLD})')
axes[1].axhline(-BEST_THRESHOLD, color='green', linestyle='--', alpha=0.7, label=f'Long entry (-{BEST_THRESHOLD})')
axes[1].axhline(0, color='black', linestyle='-', linewidth=0.5, label='Exit (0)')
axes[1].fill_between(zscore_full.index, -BEST_THRESHOLD, BEST_THRESHOLD, alpha=0.05, color='gray')
add_event_lines(axes[1])
axes[1].set_title(f'Z-Score (Lookback = {BEST_LOOKBACK} days)', fontsize=13)
axes[1].set_ylabel('Z-Score')
axes[1].set_xlabel('Date')
axes[1].set_ylim(-5, 5)
axes[1].legend(fontsize=8, ncol=2)

plt.tight_layout()
plt.show()
```

## 4.5 Volatility filtering

<div style="background-color: #e8f5e9; border-left: 4px solid #2e7d32; padding: 10px; margin: 10px 0;">
<b>TD Reference (Tutorial #2, Problem 1: GARCH estimation)</b><br>
In TD2 we showed that Brent returns exhibit strong volatility clustering (GARCH effects).
The volatility filter exploits this by blocking new trade entries when realized volatility is
abnormally high. The idea is simple: during a volatility storm (COVID crash, Ukraine war),
spreads can blow out far beyond normal z-score levels and stay there for weeks. Entering a
mean-reversion trade during such periods is dangerous. We use a 30-day rolling volatility
and block entries when it exceeds 1.5x the long-run average (following Palazzi's defaults).
</div>

```python
# ============================================================
# Volatility filter: block entries when vol is too high
# Palazzi GitHub defaults: vol_lookback=30, vol_threshold=1.5
# ============================================================
VOL_LOOKBACK = 30
VOL_THRESHOLD = 1.5

# Compute rolling volatility of the spread
spread_ret_full = spread_full.diff().dropna()
rolling_vol = spread_ret_full.rolling(VOL_LOOKBACK).std()
vol_mean = rolling_vol.expanding().mean()  # expanding average vol

# Volatility filter: True = trading allowed, False = blocked
vol_filter = rolling_vol <= VOL_THRESHOLD * vol_mean
vol_filter = vol_filter.reindex(zscore_full.index).fillna(True)

pct_blocked = (~vol_filter).sum() / len(vol_filter) * 100
print(f'Volatility filter (Palazzi defaults):')
print(f'  Lookback: {VOL_LOOKBACK} days')
print(f'  Threshold: {VOL_THRESHOLD}x average vol')
print(f'  Days blocked: {(~vol_filter).sum()} / {len(vol_filter)} ({pct_blocked:.1f}%)')

# Visualize
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(rolling_vol, color='#c62828', linewidth=0.8, label='30-day rolling vol')
ax.plot(VOL_THRESHOLD * vol_mean, color='black', linestyle='--', linewidth=0.8,
        label=f'{VOL_THRESHOLD}x avg vol (threshold)')
ax.fill_between(rolling_vol.index, 0, rolling_vol.max(),
                where=~vol_filter.reindex(rolling_vol.index).fillna(True),
                alpha=0.15, color='red', label='Blocked periods')
add_event_lines(ax)
ax.set_title('Volatility Filter: Blocked Periods (red shading)', fontsize=13)
ax.set_ylabel('Spread Volatility')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## 4.6 Adaptive trailing stop-loss

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 3.3) & GitHub</b><br>
The adaptive trailing stop-loss tracks the peak P&L of each trade and exits when the P&L
drops by more than <code>trailing_stop_factor = 0.025</code> (2.5%) from the peak. Combined with
the <code>min_holding_period = 5</code> days, this prevents premature exits while still protecting
against large drawdowns within a trade. This is one of the key innovations of the Palazzi
methodology: it turns a basic pairs strategy into a risk-managed one.
</div>

```python
# ============================================================
# FULL STRATEGY: z-score signals + vol filter + trailing stop
# + min holding period
# This is the Palazzi (2025) optimized strategy
# ============================================================
TRAILING_STOP = 0.025   # 2.5% from peak PnL (Palazzi GitHub default)
MIN_HOLD = 5            # minimum 5 days before exit allowed

# Align all series
common_idx = zscore_full.index.intersection(vol_filter.index)
z = zscore_full.loc[common_idx]
vf = vol_filter.loc[common_idx]
sp = spread_full.reindex(common_idx)
sp_ret = sp.diff()

# Generate positions with all filters
position = pd.Series(0.0, index=common_idx)
trade_pnl = 0.0          # cumulative PnL of current trade
peak_pnl = 0.0           # peak PnL of current trade
days_in_trade = 0         # counter for min holding period
entry_spread = 0.0        # spread level at entry

for i in range(1, len(common_idx)):
    prev_pos = position.iloc[i-1]
    z_val = z.iloc[i]
    vol_ok = vf.iloc[i]
    spread_change = sp_ret.iloc[i] if not np.isnan(sp_ret.iloc[i]) else 0.0

    if prev_pos == 0:
        # No position: look for entry signal
        if vol_ok:  # only enter if vol filter allows
            if z_val < -BEST_THRESHOLD:
                position.iloc[i] = 1.0    # long spread
                entry_spread = sp.iloc[i]
                trade_pnl = 0.0
                peak_pnl = 0.0
                days_in_trade = 0
            elif z_val > BEST_THRESHOLD:
                position.iloc[i] = -1.0   # short spread
                entry_spread = sp.iloc[i]
                trade_pnl = 0.0
                peak_pnl = 0.0
                days_in_trade = 0
            else:
                position.iloc[i] = 0.0
        else:
            position.iloc[i] = 0.0  # vol too high, stay flat

    else:
        # In a position: check exit conditions
        days_in_trade += 1
        trade_pnl += prev_pos * spread_change
        peak_pnl = max(peak_pnl, trade_pnl)

        # Check trailing stop (only after min holding period)
        stop_triggered = False
        if days_in_trade >= MIN_HOLD and peak_pnl > 0:
            if (peak_pnl - trade_pnl) > TRAILING_STOP:
                stop_triggered = True

        # Check mean-reversion exit (z-score crosses 0)
        mean_revert_exit = False
        if days_in_trade >= MIN_HOLD:
            if prev_pos == 1.0 and z_val >= 0:
                mean_revert_exit = True
            elif prev_pos == -1.0 and z_val <= 0:
                mean_revert_exit = True

        if stop_triggered or mean_revert_exit:
            position.iloc[i] = 0.0  # exit
        else:
            position.iloc[i] = prev_pos  # hold

# Compute strategy returns
strat_returns = (position.shift(1) * sp_ret).dropna()

# Apply transaction costs on position changes
trades = position.diff().abs()
tc_cost = trades.shift(1).reindex(strat_returns.index).fillna(0) * TC
strat_returns = strat_returns - tc_cost

# Count trades
n_trades = (trades > 0).sum()
n_entries = (trades == 1).sum() + (trades == 2).sum()

print(f'=== OPTIMIZED STRATEGY SUMMARY ===')
print(f'Pair: Brent vs {best_ticker}')
print(f'Lookback: {BEST_LOOKBACK} days | Threshold: {BEST_THRESHOLD}')
print(f'Vol filter: {VOL_LOOKBACK}d / {VOL_THRESHOLD}x | Trailing stop: {TRAILING_STOP*100:.1f}%')
print(f'Min holding period: {MIN_HOLD} days')
print(f'Transaction cost: {TC_BPS} bps per trade')
print(f'Total position changes: {int(n_trades)}')
```

```python
# ============================================================
# Split strategy returns into in-sample and out-of-sample
# ============================================================
strat_ret_is = strat_returns.loc[strat_returns.index < SPLIT_DATE]
strat_ret_oos = strat_returns.loc[strat_returns.index >= SPLIT_DATE]

# Also compute buy-and-hold Brent returns for comparison
bh_returns = returns['BZ=F'].reindex(strat_returns.index)
bh_ret_is = bh_returns.loc[bh_returns.index < SPLIT_DATE]
bh_ret_oos = bh_returns.loc[bh_returns.index >= SPLIT_DATE]

def calc_metrics(ret, label=''):
    ann_ret = ret.mean() * 252 * 100
    ann_vol = ret.std() * np.sqrt(252) * 100
    sharpe = (ret.mean() / ret.std()) * np.sqrt(252) if ret.std() > 0 else 0
    sortino_denom = ret[ret < 0].std() * np.sqrt(252)
    sortino = (ret.mean() * np.sqrt(252)) / sortino_denom if sortino_denom > 0 else 0
    cum = (1 + ret).cumprod()
    peak = cum.cummax()
    dd = (cum - peak) / peak
    max_dd = dd.min() * 100
    calmar = ann_ret / abs(max_dd) if max_dd != 0 else 0
    return {
        'Period': label,
        'Ann. Return (%)': round(ann_ret, 2),
        'Ann. Vol (%)': round(ann_vol, 2),
        'Sharpe': round(sharpe, 3),
        'Sortino': round(sortino, 3),
        'Max DD (%)': round(max_dd, 2),
        'Calmar': round(calmar, 3)
    }

metrics_is = calc_metrics(strat_ret_is, 'Optimized (IS)')
metrics_oos = calc_metrics(strat_ret_oos, 'Optimized (OOS)')
metrics_bh_is = calc_metrics(bh_ret_is, 'Buy&Hold (IS)')
metrics_bh_oos = calc_metrics(bh_ret_oos, 'Buy&Hold (OOS)')

metrics_df = pd.DataFrame([metrics_is, metrics_oos, metrics_bh_is, metrics_bh_oos]).set_index('Period')
print('Performance Metrics: Optimized Strategy vs Buy-and-Hold')
print(metrics_df.to_string())
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The key metric to focus on is the out-of-sample (OOS) performance, because in-sample results
are always optimistically biased by the parameter optimization. If the OOS Sharpe ratio is
positive and the strategy outperforms buy-and-hold on a risk-adjusted basis, that is a strong
result. Even if the OOS return is lower than in-sample (which is normal due to overfitting),
a positive Sharpe with controlled drawdown shows that the cointegration relationship holds
and the risk management (vol filter + trailing stop) is doing its job. The Calmar ratio
(return / max drawdown) is particularly important for the professor, as it measures how much
pain you endure per unit of return.
</div>

```python
# ============================================================
# Cumulative returns plot: strategy vs buy-and-hold
# ============================================================
cum_strat = (1 + strat_returns).cumprod()
cum_bh = (1 + bh_returns).cumprod()

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(cum_strat, color='#2e7d32', linewidth=1.5, label='Optimized Pairs Strategy')
ax.plot(cum_bh, color='#1565c0', linewidth=1.2, linestyle='--', label='Buy & Hold Brent')
ax.axhline(1, color='gray', linewidth=0.5)
add_event_lines(ax)
ax.set_title(f'Cumulative Returns: Optimized Strategy vs Buy & Hold', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Cumulative Return (1 = initial)')
ax.legend(fontsize=10)
ax.set_xlim(strat_returns.index[0], strat_returns.index[-1])
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# Drawdown plot
# ============================================================
cum = (1 + strat_returns).cumprod()
peak = cum.cummax()
drawdown = (cum - peak) / peak * 100

fig, ax = plt.subplots(figsize=(14, 4))
ax.fill_between(drawdown.index, drawdown, 0, color='#c62828', alpha=0.4)
ax.plot(drawdown, color='#c62828', linewidth=0.8)
add_event_lines(ax)
ax.set_title('Optimized Strategy: Drawdown (%)', fontsize=13)
ax.set_ylabel('Drawdown (%)')
ax.set_xlabel('Date')
plt.tight_layout()
plt.show()

print(f'Max drawdown: {drawdown.min():.2f}%')
print(f'Date of max drawdown: {drawdown.idxmin().strftime("%Y-%m-%d")}')
```

```python
# ============================================================
# Position plot: when are we long, short, or flat?
# ============================================================
fig, axes = plt.subplots(2, 1, figsize=(14, 7), sharex=True)

# Top: z-score with entry/exit zones
axes[0].plot(z, color='navy', linewidth=0.6)
axes[0].axhline(BEST_THRESHOLD, color='red', linestyle='--', alpha=0.5)
axes[0].axhline(-BEST_THRESHOLD, color='green', linestyle='--', alpha=0.5)
axes[0].axhline(0, color='black', linewidth=0.5)
add_event_lines(axes[0])
axes[0].set_title('Z-Score and Trading Signals', fontsize=13)
axes[0].set_ylabel('Z-Score')

# Bottom: position
axes[1].fill_between(position.index, position, 0,
                      where=position > 0, color='green', alpha=0.4, label='Long spread')
axes[1].fill_between(position.index, position, 0,
                      where=position < 0, color='red', alpha=0.4, label='Short spread')
axes[1].axhline(0, color='black', linewidth=0.5)
add_event_lines(axes[1])
axes[1].set_title('Position Over Time', fontsize=13)
axes[1].set_ylabel('Position (-1, 0, +1)')
axes[1].set_xlabel('Date')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.show()

# Position stats
pct_long = (position > 0).sum() / len(position) * 100
pct_short = (position < 0).sum() / len(position) * 100
pct_flat = (position == 0).sum() / len(position) * 100
print(f'Time in market: Long {pct_long:.1f}% | Short {pct_short:.1f}% | Flat {pct_flat:.1f}%')
```

<div style="background-color: #e8f5e9; border-left: 4px solid #2e7d32; padding: 10px; margin: 10px 0;">
<b>TD Reference (Tutorial #5: Regime-Switching Models)</b><br>
In TD5 we studied the Huisman & Mahieu (2003) three-regime model for energy prices
(normal, spike-up, spike-down). The position plot shows similar regime behavior in our
strategy: periods of active trading alternate with flat periods (when the vol filter blocks
entries or the spread is inside the no-trade zone). The COVID crash and Ukraine war periods
should show the vol filter working correctly, keeping us out of the market during the most
dangerous regimes.
</div>

```python
print('=== SECTION 4 COMPLETE ===')
print(f'Strategy: Optimized Pairs Trading (Palazzi 2025 methodology)')
print(f'Pair: Brent vs {best_ticker}')
print(f'Parameters: lookback={BEST_LOOKBACK}, threshold={BEST_THRESHOLD},')
print(f'  trailing_stop={TRAILING_STOP}, min_hold={MIN_HOLD}, vol_filter={VOL_LOOKBACK}d/{VOL_THRESHOLD}x')
print(f'Transaction cost: {TC_BPS} bps')
print(f'\nReady for Section 5: Additional Strategies (benchmarks).')
```

# 5. Additional Strategies (Benchmarks)

To properly evaluate our optimized pairs trading strategy, we need benchmarks beyond
simple buy-and-hold. The project guidelines (Tutorial 1, Slide 8) ask us to compare against
passive strategies. We implement two classic approaches that are commonly used in energy
trading: a momentum strategy (SMA crossover) and a mean-reversion strategy (Bollinger Bands).
These give us three points of comparison: passive (buy-and-hold), trend-following (SMA),
and mean-reversion on a single asset (Bollinger). Our optimized pairs strategy should ideally
beat all three on a risk-adjusted basis.

<div style="background-color: #f3e5f5; border-left: 4px solid #7b1fa2; padding: 10px; margin: 10px 0;">
<b>Paper Reference (Palazzi 2025, Section 4)</b><br>
Palazzi compares his optimized pairs strategy against: (1) buy-and-hold, (2) a simple
(non-optimized) pairs strategy without vol filter or trailing stop, and (3) a momentum strategy.
We follow the same comparison framework and add Bollinger Bands as a fourth benchmark since
it represents single-asset mean-reversion, which contrasts nicely with our pair-based approach.
</div>

## 5.1 Simple pairs trading (no optimization)

Before evaluating external benchmarks, we first compare against a **simple pairs strategy**
that uses the same pair and hedge ratio but without the Palazzi enhancements (no lookback
optimization, no vol filter, no trailing stop, no min holding period). This isolates the
added value of the optimization layer.

```python
# ============================================================
# Simple pairs strategy: fixed lookback (60 days), fixed
# threshold (2.0), no vol filter, no trailing stop
# This is the "naive" version of pairs trading
# ============================================================
SIMPLE_LOOKBACK = 60
SIMPLE_THRESHOLD = 2.0

roll_mean_s = spread_full.rolling(SIMPLE_LOOKBACK).mean()
roll_std_s = spread_full.rolling(SIMPLE_LOOKBACK).std()
zscore_simple = ((spread_full - roll_mean_s) / roll_std_s).dropna()

position_simple = pd.Series(0.0, index=zscore_simple.index)
for i in range(1, len(zscore_simple)):
    prev = position_simple.iloc[i-1]
    z_val = zscore_simple.iloc[i]

    if prev == 0:
        if z_val < -SIMPLE_THRESHOLD:
            position_simple.iloc[i] = 1.0
        elif z_val > SIMPLE_THRESHOLD:
            position_simple.iloc[i] = -1.0
        else:
            position_simple.iloc[i] = 0.0
    elif prev == 1.0:
        position_simple.iloc[i] = 0.0 if z_val >= 0 else 1.0
    elif prev == -1.0:
        position_simple.iloc[i] = 0.0 if z_val <= 0 else -1.0

sp_ret_simple = spread_full.reindex(position_simple.index).diff()
simple_ret = (position_simple.shift(1) * sp_ret_simple).dropna()
trades_simple = position_simple.diff().abs()
simple_ret = simple_ret - trades_simple.shift(1).reindex(simple_ret.index).fillna(0) * TC

print(f'Simple pairs strategy: lookback={SIMPLE_LOOKBACK}, threshold={SIMPLE_THRESHOLD}')
print(f'No vol filter, no trailing stop, no min holding period')
print(f'Trades: {int((trades_simple > 0).sum())}')
```

## 5.2 Momentum: SMA crossover on Brent

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 1, Slides 20-24)</b><br>
The SMA crossover is a classic trend-following strategy. When the short-term moving average
(50 days) crosses above the long-term one (200 days), it signals a bullish trend ("golden cross").
The reverse is a "death cross". This is pure momentum: we bet that trends persist. It is the
opposite philosophy from pairs trading (which bets on mean-reversion), so it provides a good
contrasting benchmark.
</div>

```python
# ============================================================
# SMA 50/200 crossover on Brent
# Long when SMA50 > SMA200, flat otherwise
# ============================================================
brent_price = df['BZ=F']
sma_50 = brent_price.rolling(50).mean()
sma_200 = brent_price.rolling(200).mean()

# Position: long when SMA50 > SMA200 (bullish trend)
position_sma = pd.Series(0.0, index=brent_price.index)
position_sma[sma_50 > sma_200] = 1.0
position_sma = position_sma.reindex(returns.index)

sma_ret = (position_sma.shift(1) * returns['BZ=F']).dropna()
trades_sma = position_sma.diff().abs()
sma_ret = sma_ret - trades_sma.shift(1).reindex(sma_ret.index).fillna(0) * TC

# Plot
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(brent_price, color='#1565c0', linewidth=1, label='Brent', alpha=0.7)
ax.plot(sma_50, color='orange', linewidth=1.2, label='SMA 50')
ax.plot(sma_200, color='red', linewidth=1.2, label='SMA 200')

# Shade long periods
for i in range(1, len(position_sma)):
    if position_sma.iloc[i] == 1.0 and position_sma.iloc[i-1] == 0.0:
        start = position_sma.index[i]
    elif position_sma.iloc[i] == 0.0 and position_sma.iloc[i-1] == 1.0:
        end = position_sma.index[i]
        ax.axvspan(start, end, alpha=0.08, color='green')

add_event_lines(ax)
ax.set_title('SMA 50/200 Crossover Strategy on Brent', fontsize=14)
ax.set_ylabel('Price (USD)')
ax.legend(fontsize=9)
plt.tight_layout()
plt.show()

pct_invested = (position_sma > 0).sum() / len(position_sma) * 100
print(f'SMA strategy: invested {pct_invested:.1f}% of the time')
print(f'Trades: {int((trades_sma > 0).sum())}')
```

## 5.3 Mean-reversion: Bollinger Bands on Brent

<div style="background-color: #e3f2fd; border-left: 4px solid #1565c0; padding: 10px; margin: 10px 0;">
<b>Course Reference (Lecture 1, Slides 47-50: Mean-Reversion)</b><br>
Bollinger Bands are a single-asset mean-reversion indicator. The idea is that when prices
touch the lower band (mean minus 2 std), the asset is "oversold" and should revert up. This
is similar in spirit to our pairs strategy (both exploit mean-reversion), but operates on a
single asset rather than a spread. The comparison tells us how much value the cointegration
relationship adds over simple single-asset mean-reversion.
</div>

```python
# ============================================================
# Bollinger Bands on Brent (20-day window, 2 std)
# Buy when price < lower band, sell when price > middle band
# ============================================================
BB_WINDOW = 20
BB_STD = 2.0

bb_mid = brent_price.rolling(BB_WINDOW).mean()
bb_std = brent_price.rolling(BB_WINDOW).std()
bb_upper = bb_mid + BB_STD * bb_std
bb_lower = bb_mid - BB_STD * bb_std

position_bb = pd.Series(0.0, index=brent_price.index)
for i in range(1, len(brent_price)):
    prev = position_bb.iloc[i-1]
    price = brent_price.iloc[i]
    mid = bb_mid.iloc[i]
    low = bb_lower.iloc[i]

    if np.isnan(mid) or np.isnan(low):
        position_bb.iloc[i] = 0.0
        continue

    if prev == 0:
        if price < low:
            position_bb.iloc[i] = 1.0  # buy at lower band
        else:
            position_bb.iloc[i] = 0.0
    elif prev == 1.0:
        if price >= mid:
            position_bb.iloc[i] = 0.0  # exit at middle band
        else:
            position_bb.iloc[i] = 1.0  # hold

position_bb = position_bb.reindex(returns.index)
bb_ret = (position_bb.shift(1) * returns['BZ=F']).dropna()
trades_bb = position_bb.diff().abs()
bb_ret = bb_ret - trades_bb.shift(1).reindex(bb_ret.index).fillna(0) * TC

# Plot
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(brent_price, color='#1565c0', linewidth=1, label='Brent', alpha=0.7)
ax.plot(bb_mid, color='black', linewidth=0.8, linestyle='--', label='BB Middle')
ax.plot(bb_upper, color='gray', linewidth=0.6, linestyle=':', label='BB Upper')
ax.plot(bb_lower, color='gray', linewidth=0.6, linestyle=':', label='BB Lower')
ax.fill_between(bb_mid.index, bb_lower, bb_upper, alpha=0.05, color='blue')
add_event_lines(ax)
ax.set_title('Bollinger Bands Strategy on Brent (20d, 2 std)', fontsize=14)
ax.set_ylabel('Price (USD)')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

print(f'Bollinger strategy trades: {int((trades_bb > 0).sum())}')
```

<div style="background-color: #fff8e1; border-left: 4px solid #f57f17; padding: 10px; margin: 10px 0;">
<b>Our Analysis</b><br>
The three benchmarks give us three different angles. Buy-and-hold captures the overall trend
but suffers full drawdowns during crashes. SMA crossover tries to avoid the big drops by
following trends, but it is slow to react and generates whipsaw losses in ranging markets.
Bollinger Bands bet on short-term mean-reversion within a single asset, but without the
structural relationship that cointegration provides. Our optimized pairs strategy combines
the statistical rigor of cointegration with dynamic risk management, which should give it
an edge on risk-adjusted metrics even if its raw return might be lower.
</div>

```python
# ============================================================
# Align all strategy returns to the same index for comparison
# ============================================================
common_start = max(strat_returns.index[0], simple_ret.index[0],
                   sma_ret.index[0], bb_ret.index[0])
common_end = min(strat_returns.index[-1], simple_ret.index[-1],
                 sma_ret.index[-1], bb_ret.index[-1])

all_strats = pd.DataFrame({
    'Optimized Pairs': strat_returns.loc[common_start:common_end],
    'Simple Pairs': simple_ret.loc[common_start:common_end],
    'SMA 50/200': sma_ret.loc[common_start:common_end],
    'Bollinger Bands': bb_ret.loc[common_start:common_end],
    'Buy & Hold Brent': returns['BZ=F'].loc[common_start:common_end]
}).dropna()

print(f'All strategies aligned: {len(all_strats)} common trading days')
print(f'Period: {all_strats.index[0].strftime("%Y-%m-%d")} to {all_strats.index[-1].strftime("%Y-%m-%d")}')
```
